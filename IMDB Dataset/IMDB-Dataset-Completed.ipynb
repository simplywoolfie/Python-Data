```python
#Classifying movies based on their ratings
def rating(rating):
    if rating>=7.0:
        return "excellent"
    elif rating>=6.0:
        return "good"
    else:
        return "average"
    
data['Rating Category'] = data['Rating'].apply(rating)
data.head(10)
```

```python
#Seperate values by comma
list1=[]
for value in data['Genre']:
    list1.append(value.split(','))
```

```python
list1
```

```python
#Converting 2D list to 1D list
one_d =[]
for item in list1:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
#Create list that only shows unique categories
unique_list=[]
for item in one_d:
    if item not in unique_list:
        unique_list.append(item)
```
